```python
import sympy as sp

def resolver_sistema_no_lineal():
    G, I, B = sp.symbols('G I B', real=True)

    R_0 = 864
    G_e = 140
    E_GO = 1.44
    S_I = 0.72
    sigma = 43.2
    alpha = 20000
    rho = 0.41
    k = 432
    d_0 = 0.06
    r_1 = 0.84e-3   # Equivale a 0.84 * 10^-3
    r_2 = 0.24e-5   # Equivale a 0.24 * 10^-5

    # eq1: R_0 + G_e - (E_GO + S_I*I)*G = 0
    eq1 = sp.Eq(R_0 + G_e - (E_GO + S_I * I) * G, 0)
    
    # eq2: (B * sigma * G^2) / (alpha + G^2) - (rho - k)*I = 0
    eq2 = sp.Eq((B * sigma * G**2) / (alpha + G**2) - (rho + k) * I, 0)
    
    # eq3: (-d_0 + r_1*G - r_2*G^2)*B = 0
    eq3 = sp.Eq((-d_0 + r_1 * G - r_2 * G**2) * B, 0)

    print("Calculando soluciones...\n")
    soluciones = sp.nonlinsolve([eq1, eq2, eq3], [G, I, B])

    print("Las soluciones encontradas son:")
    
    if not soluciones:
        print("El sistema no tiene solución real o compleja.")
    else:
        for i, sol in enumerate(soluciones):
            # sol es una tupla con el orden (G, I, B)
            G_sol = sol[0]
            I_sol = sol[1]
            B_sol = sol[2]
            
            print(f"\nSolución {i + 1}:")
            print(f"  Glucosa (G) = {G_sol.evalf(5)}")
            print(f"  Insulina (I)= {I_sol.evalf(5)}")
            print(f"  Células (B) = {B_sol.evalf(5)}")

if __name__ == "__main__":
    resolver_sistema_no_lineal()
```

```python
import sympy as sp

# 1. Definir las variables simbólicas
x = sp.Symbol('x')
d0, r1, r2 = sp.symbols('d0 r1 r2')

# 2. Definir la ecuación cuadrática: ax^2 + bx + c = 0
# sp.Eq crea una igualdad matemática
ecuacion = sp.Eq(-r2*x**2 + r1*x - d0, 0)

# 3. Resolver la ecuación para 'x'
raices = sp.solve(ecuacion, x)

# 4. Imprimir los resultados
print("Ecuación:")
sp.pprint(ecuacion)

print("\nLas soluciones analíticas (en símbolos) son:")
for i, raiz in enumerate(raices, start=1):
    print(f"\nRaíz {i}:")
    sp.pprint(raiz)
```

```python
import sympy as sp

def evaluar_casos_simbolicos():
    # 1. Definir las variables de estado
    G, I, B = sp.symbols('G I B', real=True)

    # 2. Definir las constantes como símbolos (sin valores numéricos)
    R_0, G_e, E_GO, S_I, sigma, alpha, rho, k, d_0, r_1, r_2 = sp.symbols(
        'R_0 G_e E_GO S_I sigma alpha rho k d_0 r_1 r_2', real=True
    )

    # 3. Definir las expresiones del sistema (lado derecho de las ecuaciones)
    dG_dt = R_0 + G_e - (E_GO + S_I * I) * G
    dI_dt = (B * sigma * G**2) / (alpha + G**2) - (rho - k) * I
    dB_dt = (-d_0 + r_1 * G - r_2 * G**2) * B

    # Función auxiliar para imprimir los resultados de forma ordenada
    def imprimir_caso(nombre_caso, sustitucion):
        print(f"--- Sistema cuando {nombre_caso} ---")
        print(f"dG/dt = {sp.simplify(dG_dt.subs(sustitucion))}")
        print(f"dI/dt = {sp.simplify(dI_dt.subs(sustitucion))}")
        print(f"dβ/dt = {sp.simplify(dB_dt.subs(sustitucion))}\n")

    # 4. Evaluar y mostrar los tres casos por separado
    
    # Caso 1: Solo G = 0
    imprimir_caso("G = 0", {G: 0})

    # Caso 2: Solo I = 0
    imprimir_caso("I = 0", {I: 0})

    # Caso 3: Solo B = 0
    imprimir_caso("B = 0", {B: 0})

if __name__ == "__main__":
    evaluar_casos_simbolicos()
```
